```python
X = df['MARKS'].values.reshape(-1,1)
```

```python
Y = df['RANK'].values.reshape(-1,1)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
regressor = GradientBoostingRegressor(n_estimators=5000, learning_rate=0.02)
model = regressor.fit(X,Y)
```
